```python
df = pd.read_csv('train_df.csv', encoding='utf-8')
y = pd.read_csv('train_y.csv', encoding='utf-8')

# dfy_d = pd.DataFrame(y.석식계)
# dfy_l = pd.DataFrame(y.중식계)

# dfx_l = sm.add_constant(df.drop(columns=['월저녁평균', '요일저녁평균']))
# dfx_d = sm.add_constant(df.drop(columns=['월점심평균', '요일점심평균']))

df_l = pd.concat([df.drop(columns=['월저녁평균', '요일저녁평균']), y.중식계], axis=1)
df_l.columns = ['yoil', 'holiday', 'length', 'work', 'vac', 'out', 'night', 'home' , 'year', 'month', 'day', 'week', 'rain', 'discom', 'temp', 'wla', 'mla',  'corona', 'lunch']
df_l = sm.add_constant(df_l)

df_d = pd.concat([df.drop(columns=['월점심평균', '요일점심평균']), y.석식계], axis=1)
df_d.columns = ['yoil', 'holiday', 'length', 'work', 'vac', 'out', 'night', 'home' , 'year', 'month', 'day', 'week', 'rain', 'discom', 'temp', 'wda', 'mda',  'corona', 'dinner']
df_d = sm.add_constant(df_d)
# model_l = sm.OLS(dfy_l, dfx_l)
# result_l = model_l.fit()

# model_d = sm.OLS(dfy_d, dfx_d)
# result_d = model_d.fit()

model_l = sm.OLS.from_formula('lunch ~ C(holiday) +  vac + out + night + year + day + rain + discom + wla + mla + corona', df_l)
result_l = model_l.fit()

model_d = sm.OLS.from_formula('dinner ~ C(holiday) +  vac + out + night + year + day +  mda', df_d)
result_d = model_d.fit()
```
