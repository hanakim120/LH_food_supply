```python
import re

for i in range(len(data)):
    data[i] = data[i].rstrip()
    data[i] = re.sub('▁', '', data[i])
data
```

```python
from nltk import word_tokenize

vocabs = np.unique(word_tokenize(' '.join(data)))
len(vocabs)
```

```python
new_vocabs = []
for i in range(len(vocabs)):
    new_vocabs.append((len(vocabs[i]), vocabs[i]))
```

```python
new_vocabs.sort(reverse=True)
```

```python
new_vocabs = [x[1] for x in new_vocabs]
```

```python
pd.DataFrame(new_vocabs, columns=['menu']).to_csv('menu.sub.csv', index=False, encoding='utf-8-sig')
```

```python
menu = pd.read_csv('./data/menu.sub.csv', encoding='utf-8')
menu_list = list(menu.menu)
```

```python
with open('./data/corpus.sub.txt', 'r', -1, encoding='utf-8') as f :
    data = f.readlines()

for i in range(len(data)) :
    data[i] = data[i].rstrip()
    data[i] = re.sub('▁', '', data[i])

array = [d.split() for d in data]

result = pd.DataFrame(np.zeros((len(array), menu.shape[1] - 1)))

for i in range(len(array)):
    for j in range(len(array[i])):
        if array[i][j] in menu_list:
            result.loc[i, :] += menu[menu['menu'] == array[i][j]].values.reshape(-1)[1:]
```

```python
result.to_csv('./data/embedding.csv', index=False)
```
