```python
data = pd.read_csv('data.csv', encoding='utf-8')
data['const'] = 1.
target = pd.read_csv('target.csv', encoding='utf-8')
test = pd.read_csv('test.csv', encoding='utf-8')
test['const'] = 1.
```

```python
sample = pd.read_csv('./data/sample_submission.csv', encoding='utf-8')
```

```python
model = sm.OLS(target.석식계, data)
result = model.fit()
result.summary()
```

```python
influence = result.get_influence()
cooks_d2, pvals = influence.cooks_distance
k = influence.k_vars
fox_cr = 4 / (len(target) - k - 1)

idx = np.where(cooks_d2 > fox_cr)[0]
print(idx)

new_idx = list(set(range(len(data))).difference(idx))
dinner_data = data.iloc[new_idx, :].reset_index(drop=True)
dinner_target = target.iloc[new_idx, :].reset_index(drop=True)
data, target
```

```python
dinner_drop_col = ['요일', '공휴일길이', '출근', 'month', '강수량', '불쾌지수', '체감온도', '미세먼지', '요일점심평균', '월점심평균', '주점심평균', '주저녁평균', 'breakfast_0', 'lunch_1']
model = sm.OLS(dinner_target.석식계, dinner_data.drop(columns=dinner_drop_col))
result = model.fit()
result.summary()
```

```python
sample.석식계 = result.predict(test.drop(columns=dinner_drop_col))
```

```python
model = sm.OLS(target.중식계, data)
result = model.fit()
result.summary()
```

```python
influence = result.get_influence()
cooks_d2, pvals = influence.cooks_distance
k = influence.k_vars
fox_cr = 4 / (len(target) - k - 1)

idx = np.where(cooks_d2 > fox_cr)[0]
print(idx)

new_idx = list(set(range(len(data))).difference(idx))
lunch_data = data.iloc[new_idx, :].reset_index(drop=True)
lunch_target = target.iloc[new_idx, :].reset_index(drop=True)
```

```python
lunch_drop_col = ['공휴일길이', 'month', '불쾌지수', '체감온도', '폭염', '미세먼지', '요일저녁평균', '주점심평균', '주저녁평균', '확진자수', 'lunch_0', 'lunch_2', '사망자수', 'week', 'day']
model = sm.OLS(lunch_target.중식계, lunch_data.drop(columns=lunch_drop_col))
result = model.fit()
result.summary()
```

```python
sample.중식계 = result.predict(test.drop(columns=lunch_drop_col))
```

```python
sample.to_csv('submission_sm.csv', index=False)
```
